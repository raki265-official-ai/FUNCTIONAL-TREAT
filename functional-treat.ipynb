## FUNCTIONAL TREAT🎯

```python
data =[]

def input_data():
    global data

    user_input = input("\nEnter data for a 1D array (separated by spaces):\n")

    data = list(map(int, user_input.split()))

    print("\nData has been sorted successfully!")





def display_summary():
    global data


    if len(data) == 0:
        print("\nNo data available. Please input data first.")
        return


    print("\nData Summary:")
    print(f"- Total elements: {len(data)}")
    print(f"- Minimum value: {min(data)}")
    print(f"- Maximum value: {max(data)}")
    print(f"- Sum of all values:{sum(data)}")
    print(f"- Average value: {sum(data)/ len(data):.2f}")



def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n* factorial(n-1)


def calculate_factorial():

    number = int(input("\nEnter a number to calculate its factorial: "))

    if number < 0:
        print("Factorial is not defined for negative numbers.")
    else:
        print(f"\nFactorial of {number} is: {factorial(number)}")


def filter_data():
    global data

    if len(data) == 0:
        print("\nNo data available. Please input data first.")
        return

    threshold = int(input("\nEnter a threshold value to filter out data above this value"))

    filtered = list(filter(lambda x: x >= threshold, data))

    print(f"\nFiltered data (values >={threshold}):")
    print(*filtered, sep=", ")

def sort_data():
    global data

    if len(data) ==0:
        print("\nNo data available. Please input data first.")
        return

    print("\nChoose soring option:")
    print("1. Ascending")
    print("2. Descending")

    choice = input("\nEnter your choice: ")

    if choice == "1":
        sorted_data = sorted(data)

        print("\nSorted data in Ascending order:")
        print(*sorted_data,sep=", ")

    elif choice =="2":
        sorted_data = sorted(data, reverse=True)

        print("\nSorted data in Descending order:")
        print(*sorted_data,sep=", ")

    else:
        print("Invalid choice!")


def dataset_statistics():
    global data

    minimum = min(data)
    maximum = max(data)
    total = sum(data)
    average = total / len(data)

    return minimum, maximum, total, average



def display_statistics():
    global data

    if len(data) == 0:
        print("\nNo data available. Please input data first.")

    minimum, maximum, total, average = dataset_statistics()

    print("\nDataset Statistcs:")
    print(f"- Minimum value:{minimum}")
    print(f"- Maximum value:{maximum}")
    print(f"- Sum of all values:{total}")
    print(f"- Average value: {average:.2f}")



def main():
    while True:

        print("\n==================================")
        print("Welcome to the data Analyzer and Transformer program")
        print("=====================================")


        print("\nMain Menu:")
        print("1. Input data:")
        print("2. Display data summary (Built-in Functions)")
        print("3. Calculate Factorial (Recursion)")
        print("4. Filter Data by Threshold (Lambda Function)")
        print("5. Sort Data")
        print("6. Display Dataset Statistics (Return Multiple Values)")
        print("7. Exit Program")

        choice = input("\nPlease enter your choice: ")

        if choice =="1":
            input_data()

        elif choice =="2":
            display_summary()

        elif choice =="3":
            calculate_factorial()

        elif choice =="4":
            filter_data()

        elif choice =="5":
            sort_data()

        elif choice =="6":
            display_statistics()

        elif choice =="7":
            print("\nThank you for using the Data Analyzer and Transformer program. Goodbye!")
            break
        else:
            print("\nInvalid choice! Please try again.")


main()
        
```
